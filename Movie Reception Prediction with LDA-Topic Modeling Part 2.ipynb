### Reviews with Topic Distribution for Model18

```python
corpus_transformed = model_18[corpus_BoW]
corpus_transformed
```

```python
#Top topics for each Review
distr_corpus = [corpus_transformed[i] for i in range(len(corpus_transformed))]
distr_corpus[:1]
```

```python
reviews = review_texts.values.tolist()
```

```python
# Topic Distribution by Review List
q=list(zip(distr_corpus,reviews ))
```

```python
topic_reviews = pd.DataFrame(q, columns=['Topic Distribution', 'Review'])
topic_reviews.head()
```

```python
#Save TopicDistribution-Review DF 
topic_reviews.to_pickle('pickled/topicDistr_reviews.pkl')
```
